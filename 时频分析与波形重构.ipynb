```python
import os
import pycwb
from pycwb.config import Config
from pycwb.modules.logger import logger_init

# æ£€æŸ¥ç¯å¢ƒå˜é‡ 'HOME_WAT_FILTERS' æ˜¯å¦å·²è®¾ç½®ï¼Œå¦‚æœæ²¡æœ‰è®¾ç½®ï¼Œåˆ™è®¾ç½®ä¸º pycwb åº“çš„ vendor æ–‡ä»¶å¤¹è·¯å¾„
if not os.environ.get('HOME_WAT_FILTERS'):
    # è·å– pycwb åº“çš„å®‰è£…è·¯å¾„
    pyburst_path = os.path.dirname(os.path.abspath(pycwb.__file__))
    # è®¾ç½®ç¯å¢ƒå˜é‡ HOME_WAT_FILTERSï¼ŒæŒ‡å‘ pycwb çš„ vendor ç›®å½•
    os.environ['HOME_WAT_FILTERS'] = f"{os.path.abspath(pyburst_path)}/vendor"

# åˆå§‹åŒ–æ—¥å¿—è®°å½•å™¨ï¼Œé€šå¸¸ç”¨äºè®°å½•åº”ç”¨çš„è¿è¡Œæ—¥å¿—
logger_init()

# ä½¿ç”¨ drive_path åˆ›å»ºé…ç½®å¯¹è±¡
config = Config(drive_path)
```

```python
import os
import requests
from gwosc.locate import get_urls
from gwpy.timeseries import TimeSeries

# GW190521_074359 çš„ GPS è§¦å‘æ—¶é—´
t0 = 1242442967.4
ifo_list = ["H1", "L1"]

data = []
for ifo in ifo_list:
    # è·å–ä¸è§¦å‘æ—¶é—´ t0 ç›¸å…³çš„ URL åˆ—è¡¨
    urls = get_urls(ifo, t0, t0)

    # å¦‚æœæ²¡æœ‰æ‰¾åˆ°æ•°æ®ï¼Œè·³è¿‡è¯¥å¹²æ¶‰ä»ª
    if not urls:
        print(f"No data found for {ifo} at t0 = {t0}")
        continue

    # é€‰æ‹©åˆ—è¡¨ä¸­çš„æœ€åä¸€ä¸ª URLï¼ˆé€šå¸¸ä¸ºæœ€æ–°æ•°æ®ï¼‰
    url = urls[-1]
    print('Downloading:', url)

    # è·å–æ–‡ä»¶åå¹¶ä¸‹è½½æ•°æ®æ–‡ä»¶
    fn = os.path.basename(url)
    with open(fn, 'wb') as strainfile:
        strainfile.write(requests.get(url).content)

    # ä½¿ç”¨ gwpy çš„ TimeSeries æ¥è¯»å–å¹¶å¤„ç†æ•°æ®
    strain = TimeSeries.read(fn, format='hdf5.gwosc')

    # å°†æ•°æ®è£å‰ªåˆ° t0 å‰å 150 ç§’èŒƒå›´å†…
    d = strain.crop(t0 - 150, t0 + 150)

    # é‡æ–°é‡‡æ ·æ•°æ®åˆ° 2048 Hz
    d_resampled = d.resample(2048)

    # å°†å¤„ç†åçš„æ•°æ®æ·»åŠ åˆ°åˆ—è¡¨ä¸­
    data.append(d_resampled)
```

```python
from pycwb.modules.data_conditioning import data_conditioning
from pycwb.modules.plot import plot_spectrogram

# å¯¹æ•°æ®è¿›è¡Œé¢„å¤„ç†ï¼ˆæ»¤æ³¢ã€å½’ä¸€åŒ–ç­‰ï¼‰
# è¿”å›å¤„ç†åçš„æ•°æ® strains å’Œå™ªå£° RMSï¼ˆå‡æ–¹æ ¹ï¼‰å€¼ nRMS
strains, nRMS = data_conditioning(config, data)

# è®© matplotlib å›¾è¡¨åµŒå…¥åœ¨ notebook ä¸­æ˜¾ç¤º
%matplotlib inline

# ç»˜åˆ¶ç¬¬ä¸€ä¸ªå¹²æ¶‰ä»ªï¼ˆå¦‚ L1ï¼‰çš„é¢‘è°±å›¾
plt = plot_spectrogram(strains[0], gwpy_plot=True)

# è·å–å½“å‰å›¾çš„åæ ‡è½´ï¼Œå¹¶è®¾ç½®é¢‘ç‡æ˜¾ç¤ºèŒƒå›´ä¸º 15 Hz åˆ° 1024 Hz
ax = plt.gca()
ax.set_ylim(15,1024)

# è¾“å‡ºå¤„ç†åçš„æ•°æ® strains å’Œå™ªå£°æŒ‡æ ‡ nRMS
strains, nRMS
```

```python
from pycwb.modules.coherence import coherence

# è®¡ç®—å„å¹²æ¶‰ä»ªä¹‹é—´çš„ä¿¡å·ç›¸å¹²æ€§ï¼Œæå–å€™é€‰äº‹ä»¶ç‰‡æ®µ
fragment_clusters = coherence(config, strains, nRMS)
```

```python
from pycwb.modules.super_cluster import supercluster
from pycwb.types.network import Network

# åˆ›å»ºç½‘ç»œå¯¹è±¡ï¼Œè¡¨ç¤ºå¤šä¸ªå¹²æ¶‰ä»ªååŒåˆ†æçš„ç½‘ç»œ
network = Network(config, strains, nRMS)

# åŸºäºç›¸å¹²äº‹ä»¶ç‰‡æ®µè¿›è¡Œèšç±»ï¼Œç”Ÿæˆå€™é€‰äº‹ä»¶åˆ—è¡¨
pwc_list = supercluster(config, network, fragment_clusters, strains)
```

```python
%matplotlib inline
from gwpy.spectrogram import Spectrogram

# éå†ç¬¬ä¸€ä¸ªå€™é€‰äº‹ä»¶ä¸­çš„æ‰€æœ‰å­ç°‡
for cluster in pwc_list[0].clusters:
    # è·å–è¯¥ç°‡çš„ç¨€ç–ä¼¼ç„¶å›¾åŠå…¶æ—¶é—´/é¢‘ç‡åˆ†è¾¨ç‡ä¿¡æ¯
    merged_map, start, dt, df = cluster.get_sparse_map("likelihood")

    # ç»˜åˆ¶ä¼¼ç„¶å›¾ï¼ˆSpectrogramï¼‰
    plt = Spectrogram(merged_map, t0=start, dt=dt, f0=0, df=df).plot()
    plt.colorbar()  # æ·»åŠ é¢œè‰²æ¡ä»¥æ˜¾ç¤ºå¼ºåº¦
```

```python
from pycwb.modules.likelihood import likelihood

# å¯¹å€™é€‰äº‹ä»¶æ‰§è¡Œä¼¼ç„¶åˆ†æï¼Œæå–äº‹ä»¶ä¿¡æ¯ã€ç°‡ä¿¡æ¯å’Œå¤©ç©ºå›¾ç»Ÿè®¡ç»“æœ
events, clusters, skymap_statistics = likelihood(config, network, pwc_list)
```
